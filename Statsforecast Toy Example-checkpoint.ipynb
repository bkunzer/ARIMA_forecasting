```python
import numpy as np
import pandas as pd
import statsforecast
from statsforecast import StatsForecast
from statsforecast.models import AutoARIMA, ETS, SeasonalNaive, Naive

#matplolib used only for plotting in this notebook
import matplotlib.pyplot as plt
```

```python
def predict(data):
    '''
    Input:
        data - a 1D array (vector) [N] of N obversations over time
        (note this assumes regular sampling frequency)
                    
    Returns:
        forecast_14_day - a 1D array (vector) [N] of N predicted values over horizon time (14 days or 14 intervals)
    '''
    
    # build dataframe from data
    dataDF = pd.DataFrame(
        {
            "unique_id": np.ones(len(data)),
            "ds": pd.date_range(start="2022-01-01", periods=len(data), freq="D"),
            "y": data,
        }
    )
        
    # set parameters
    horizon = 14
    samp_period = 'D' 
    ''' calculate sample frequency from data 
        (if the input is a dataframe series [N,2] of obversations over time with the dataframe
        index set to a pandas datetime using pd.to_datetime (dtype = datetime64[ns])): 
        
        samp_period = pd.Timedelta(data.index[1] - data.index[0])
        horizon = int(pd.Timedelta(days=14).total_seconds() /
        pd.Timedelta(data.index[1] - data.index[0]).total_seconds())
    '''
    #train model 
    models = [AutoARIMA(seasonal = False)] #non-seasonal ARIMA
    model = StatsForecast(df = dataDF, models = models, freq = samp_period, n_jobs=-1)

    forecast = model.forecast(horizon).reset_index()
        
    #convert the dataframe column into a vector fro output
    forecast_14_day_vector = forecast['AutoARIMA'].to_numpy()
    
    return forecast_14_day_vector
```

```python
from statsforecast.utils import AirPassengersDF
Y_df = AirPassengersDF
Y_df.head()
```

```python
fig1, ax = plt.subplots(1, 1, figsize = (20, 7))


x = Y_df.index
y = Y_df['y']
plt.plot(x,y)

ax.set_title('AirPassengers', fontsize=22)
ax.set_ylabel('Monthly Passengers', fontsize=20)
ax.set_xlabel('Timestamp [t]', fontsize=20)
ax.grid()
```

```python
Y_train_df = Y_df[Y_df.ds<='1959-12-31'] # 132 train
Y_test_df = Y_df[Y_df.ds>'1959-12-31'] # 12 test
```

```python
predict_values = predict(Y_train_df['y'])
```

```python
fig2, ax = plt.subplots(1, 1, figsize = (20, 7))
plt.plot(predict_values, 'red')
ax.set_title('AirPassengers Forecast', fontsize=22)
ax.set_ylabel('Daily Passengers', fontsize=20)
ax.set_xlabel('Timestamp [t]', fontsize=20)
ax.grid()
```

```python
df_predict = pd.DataFrame(predict_values, columns=['values'])
plot_df = pd.concat([Y_train_df['y'], df_predict['values']])

plot_df.index = pd.RangeIndex(len(plot_df.index)) #reset the index
plot_df.index = range(len(plot_df.index))

```
